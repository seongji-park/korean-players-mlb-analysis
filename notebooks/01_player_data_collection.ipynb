```python
from pybaseball import statcast_pitcher

# 박찬호: Statcast 데이터는 2008년부터 존재
# (Statcast는 2008년부터 시작, 그 전 데이터는 없음)
# 박찬호 MLB 활동: 1994-2010 → Statcast 범위: 2008-2010

print("=== 박찬호 2008-2010 Statcast ===")
park_statcast = statcast_pitcher('2008-03-01', '2010-10-31', 120221)
print(f"박찬호 투구 수: {len(park_statcast)}")

print("\n=== 김병현 2007 Statcast ===")
kim_statcast = statcast_pitcher('2007-03-01', '2007-10-31', 218294)
print(f"김병현 투구 수: {len(kim_statcast)}")

print("\n=== 봉중근 Statcast ===")
# 봉중근은 2002-2004 활동 → Statcast 이전이라 데이터 없음
print("봉중근: 2002-2004 활동 → Statcast 범위 밖 (데이터 없음)")
```

```python
from pybaseball import statcast_batter

print("=== 추신수 2008-2020 Statcast ===")
choo_statcast = statcast_batter('2008-03-01', '2020-10-31', 425783)
print(f"추신수 타석 수: {len(choo_statcast)}")

print("\n=== 배지환 2022-2025 Statcast ===")
bae_statcast = statcast_batter('2022-03-01', '2025-10-31', 678225)
print(f"배지환 타석 수: {len(bae_statcast)}")

print("\n=== 최지만 2016-2023 Statcast ===")
choi_statcast = statcast_batter('2016-03-01', '2023-10-31', 596847)
print(f"최지만 타석 수: {len(choi_statcast)}")
```

```python
# Statcast 데이터 저장
choo_statcast.to_csv('../data/choo_statcast.csv', index=False)
bae_statcast.to_csv('../data/bae_statcast.csv', index=False)
choi_statcast.to_csv('../data/choi_statcast.csv', index=False)

if len(park_statcast) > 0:
    park_statcast.to_csv('../data/park_statcast.csv', index=False)
if len(kim_statcast) > 0:
    kim_statcast.to_csv('../data/kim_statcast.csv', index=False)

print("=== 저장 완료! ===")
print(f"추신수: {len(choo_statcast)} 투구/타석")
print(f"배지환: {len(bae_statcast)} 투구/타석")
print(f"최지만: {len(choi_statcast)} 투구/타석")
print(f"박찬호: {len(park_statcast)} 투구")
print(f"김병현: {len(kim_statcast)} 투구")
```
